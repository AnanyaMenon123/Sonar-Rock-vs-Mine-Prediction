```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/content/Copy of sonar data.csv")
```

```python
dataset #columns represent the angles, for each observation based on the transmission at an angle, it is classified as a rock or a mine the data values represent the energy within a particular frequency band between 0.0 to 0.10
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

Data Preprocessing


```python
dataset.head()
```

```python
#number of rows and columns
dataset.shape
```

```python
dataset.describe()
```

```python
x=dataset.iloc[:,:-1].values
```

```python
y=dataset.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
classifier=LogisticRegression()
```

```python
y_train
```

```python
x_train
```

```python
y
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
accuracy=accuracy_score(y_test,y_pred)
```
